```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/clean/result.csv')
affordable = df[df["IS_AFFORDABLE"] == 1]

affordable
```

## b) What is the current distribution of landlords and housing listed in current affordable housing programs?

```python
distribution = affordable["OWNER"].value_counts()
distribution.plot.pie(autopct='%.0f%%', title="Distribution of affordable housing by landlord")
distribution
```

## c) What is the geographic distribution of these landlords by city council district?

```python
distribution = affordable["CITY"].value_counts()
distribution.plot(kind='bar', title='Landlords by City Council District')
distribution
```
